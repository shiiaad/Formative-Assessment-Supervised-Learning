# Import Required Libraries

```python
import pandas as pd
import numpy as np

# Dataset
from sklearn.datasets import load_breast_cancer

# Train Test Split
from sklearn.model_selection import train_test_split

# Feature Scaling
from sklearn.preprocessing import StandardScaler

# Models
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier

# Evaluation
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

# Load Dataset
#### Loads dataset into variable data.

#### X

#### Contains input features like:

#### -mean radius
#### -mean texture
#### -mean perimeter etc.
#### y

#### Contains output labels:

#### -0 = Malignant (Cancerous)
#### -1 = Benign (Non-cancerous)

```python
data = load_breast_cancer()

X = pd.DataFrame(data.data, columns=data.feature_names)
y = pd.Series(data.target)
```

# View Dataset

#### X.head()

#### Shows first 5 rows of features.

#### y.head()

#### Shows first 5 target values.

```python
print(X.head())
print(y.head())
```

# Check Missing Values

#### This checks if any values are missing.

#### Result:

#### This dataset has no missing values.

```python
print(X.isnull().sum())
```

# Feature Scaling

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

# Train Test Split

#### test_size=0.2

#### 20% data for testing.

#### random_state=42

#### Same split every time.

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

# Create Models

```python
models = {
    "Logistic Regression": LogisticRegression(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "SVM": SVC(),
    "KNN": KNeighborsClassifier()
}
```

# Train and Evaluate Models

```python
results = {}

for name, model in models.items():
    
    # Train Model
    model.fit(X_train, y_train)
    
    # Predict
    y_pred = model.predict(X_test)
    
    # Accuracy
    acc = accuracy_score(y_test, y_pred)
    
    results[name] = acc
    
    print("Model:", name)
    print("Accuracy:", acc)
    print(confusion_matrix(y_test, y_pred))
    print(classification_report(y_test, y_pred))
    print("-"*50)
```

# Compare Results

```python
result_df = pd.DataFrame(results.items(), columns=["Model", "Accuracy"])
print(result_df.sort_values(by="Accuracy", ascending=False))
```

# Best and Worst Model

```python
best_model = result_df.loc[result_df["Accuracy"].idxmax()]
worst_model = result_df.loc[result_df["Accuracy"].idxmin()]

print("Best Model:")
print(best_model)

print("Worst Model:")
print(worst_model)
```

# Final Conclusion
### Best Model: Random Forest / SVM
### Worst Model: Decision Tree

## Because:

### Random Forest combines many trees.
### SVM handles complex boundaries.
### Decision Tree may overfit.
